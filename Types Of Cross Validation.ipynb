```python
import pandas as pd
df=pd.read_csv("C:\\Users\\ANINDITA\\Downloads\\cancer_dataset.csv")
df.head(10)
```

```python
#independent and dependent features
X=df.iloc[:,2:]
Y=df.iloc[:,1]
```

```python
X=X.dropna(axis=1)#drop unnecessay column
```

```python
X.head()
```

```python
#check balanced/unbalanced dataset
Y.value_counts()
```

```python
#HoldOut Validation Approach- Train And Test Split
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=100)
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
result = model.score(X_test, Y_test)
print(result)
```

```python
# K Fold Cross Validation
```

```python
from sklearn.model_selection import KFold
model=DecisionTreeClassifier()
kfold_validation=KFold(10)
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,X,Y,cv=kfold_validation)
print(results)
print(np.mean(results))
```

```python
# Stratified K-fold Cross Validation
```

```python
from sklearn.model_selection import StratifiedKFold
skfold=StratifiedKFold(n_splits=5)
model=DecisionTreeClassifier()
scores=cross_val_score(model,X,Y,cv=skfold)
print(scores)
print(np.mean(scores))
```

```python
# Leave One Out Cross Validation(LOOCV)
```

```python
from sklearn.model_selection import LeaveOneOut
model=DecisionTreeClassifier()
leave_validation=LeaveOneOut()
results=cross_val_score(model,X,Y,cv=leave_validation)
```

```python
print(np.mean(results))
```

```python
#Repeated Random Test-Train Splits
#This technique is a hybrid of traditional train-test splitting and the k-fold 
#cross-validation method. In this technique, we create random splits of the data in the training-test 
#set manner and then repeat the process of splitting and evaluating the algorithm multiple times, just like the cross-validation method.
```

```python
from sklearn.model_selection import ShuffleSplit
model=DecisionTreeClassifier()
ssplit=ShuffleSplit(n_splits=10,test_size=0.30)
results=cross_val_score(model,X,Y,cv=ssplit)
```
